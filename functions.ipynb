```python
import itertools
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}
docs_ = list(directories.values())
docs = list(itertools.chain(*docs_))
def main():
    user_input = input('Введите команду. Для просмотра всех доступных команд введите "help". Для возврата в предыдущее меню введите "back"\n')
    while True:
        if user_input == 'p':
            show_name()
            break
        elif user_input == 's':
            show_shelf()
            break
        elif user_input == 'l':
            show_list()
            break
        elif user_input == 'as':
            new_shelf()
            break
        elif user_input == 'd':
            delete_doc()
            break
        elif user_input == 'm':
            move_doc()
            break
        elif user_input == 'a':
            new_doc()
            break
        elif user_input == 'help':
            print('Список команд:\np – Вывод имени владельца по номеру документа\ns – Вывод номера полки по номеру документа\nl – Вывод списка всех документов\nas – Добавление новой полки в перечень полок\nd – Удаление заданного документа\nm – Перенос документа с одной полки на другую\na – Добавление нового документа')
            return main()
        elif user_input == 'back':
            break
        elif user_input == 'test':
            print(docs, directories, documents)
            return main()
        else:
            print('Вы ввели некорретную команду')
            return main()
def show_name():
    number_input = input('Введите номер документа\n')
    #docs = str(directories.values())
    if number_input in docs:
        for i in documents:
            if number_input in i.values():
                for name in i.items():
                    print(i['name'])
                    return main()
    elif number_input == 'back':
        return main()
    else:
        print('Вы ввели номер несуществующего документа')
        return show_name()           
def show_shelf():
    number_input = input('Введите номер документа\n')
    #docs = str(directories.values())
    if number_input in docs:
        for i in directories.items():
                i = list(i)
                if number_input in i[1]:
                    print('Указанный документ находится на полке', i[0])
                    return main()
    elif number_input == 'back':
        return main()
    else:
        print('Вы ввели номер несуществующего документа')
        return show_shelf()
def new_shelf():
    user_input = input('Введите номер добавляемой полки\n')
    if user_input in directories.keys():
        print('Полка уже существует')
        return new_shelf()
    elif user_input == 'back':
        return main()
    else:
        directories[user_input] = []
        print('Полка добавлена')
        return main()
def delete_doc():
    number_input = input('Введите номер удаляемого документа\n')
    #docs = str(directories.values())
    if number_input in docs:
        for i in directories.values():
            if number_input in i:
                i.remove(number_input)
        for k in documents:
            if number_input in k.values():
                documents.remove(k)
        for l in docs:
            if number_input in docs:
                docs.remove(number_input)
        print('Документ удалён')
        return main()
    elif number_input == 'back':
        return main()
    else:
        print('Вы ввели номер несуществуемого документа')
        return delete_doc()
def move_doc():
    number_input = input('Введите номер документа\n')
    #docs = str(directories.values())
    if number_input in docs:
        for i in directories.values():
            if number_input in i:
                i.remove(number_input)
                shelf_input = input('Введите номер целевой полки\n')
                if shelf_input in directories.keys():
                    for k in directories.items():
                        k = list(k)
                        if shelf_input in k:
                            k[1].append(number_input)
                    print('Документ перемещен')
                    return main()
                else:
                    print('Вы ввели номер несуществующей полки')
                    return move_doc()
    elif number_input == 'back':
        return main()
    else: 
        print('Вы ввели номер несуществуемого документа')
        return move_doc()
def new_doc():
    #docs = str(directories.values())
    type_input = input('Введите тип добавляемого документа\n')
    if type_input == 'back':
        return main()
    else:
        number_input = input('Введите номер добавляемого документа\n')
        if number_input == 'back':
            return main()
        else:
            name_input = input('Введите имя владельца добавляемого документа\n')
            if name_input == 'back':
                return main()
            else:
                shelf_input = input('Введите номер полки, в которую будет добавлен документ\n')
                if shelf_input in directories.keys():
                    i = {'type': type_input, 'number': number_input, 'name': name_input}
                    documents.append(i)
                    for k in directories.items():
                        k = list(k)
                        if shelf_input in k:
                            k[1].append(number_input)
                            docs == docs.append(number_input)
                            print('Документ добавлен')
                            return main()
                else:
                    print('Вы ввели номер несуществующей полки')
                    return new_doc()
def show_list():
    for i in documents:
        print(*i.values())
    return main()
while True:
    initial_input = input('Для начала введите "start". Для завершения программы введите "quit"\n')
    if initial_input == 'quit':
        break
    elif initial_input == 'start':
        main()
    else:
        print('Вы ввели некорректную команду')
```
