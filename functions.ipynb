```python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}
def main():
    while True:
        if user_input == 'p':
            show_name()
            break
        elif user_input == 's':
            show_shelf()
            break
        elif user_input == 'l':
            show_list()
            break
        elif user_input == 'as':
            new_shelf()
            break
        elif user_input == 'd':
            delete_doc()
            break
        elif user_input == 'm':
            move_doc()
            break
        elif user_input == 'a':
            new_doc()
            break
        elif user_input == 'help':
            print('Список команд:\np – команда, которая по номеру документа выводит имя его владельца;\ns – команда, которая по номеру документа выведит номер полки, на которой он находится\nl – команда, которая выводит список всех документов\nas – команда, которая добавляет новую полку в перечень\nd – удаление заданного документа\nm – move – перенос документа с одной полки на другую\na – добавление нового документа')
            break
def show_name():
    i = str(directories.values())
    number_input = input('Введите номер документа')
    if number_input in i:
        for k in documents:
            if number_input in k.values():
                for name in k.items():
                    return print(k['name'])
    else:
        print('Вы ввели номер несуществующего документа')
        return show_name()
def show_shelf():
    i = str(directories.values())
    number_input = input('Введите номер документа')
    if number_input in i:
        for k in directories.items():
                k = list(k)
                if number_input in k[1]:
                    return print('Указанный документ находится на полке', k[0])
    else:
        print('Вы ввели номер несуществующего документа')
        return show_shelf()
def new_shelf():
    user_input = input('Введите номер добавляемой полки ')
    if user_input in directories.keys():
        print('Полка уже существует')
        return new_shelf()
    else:
        directories[user_input] = []
        return print('Полка добавлена')
def delete_doc():
    i = str(directories.values())
    number_input = input('Введите номер удаляемого документа ')
    if number_input in i:
        for i in directories.values():
            if number_input in i:
                i.remove(number_input)
        for k in documents:
            if number_input in k.values():
                documents.remove(k)
        return print('Документ удалён')
    else:
        print('Вы ввели номер несуществуемого документа')
        return delete_doc()
def move_doc():
    i = str(directories.values())
    number_input = input('Введите номер документа')
    if number_input in i:
        for i in directories.values():
            if number_input in i:
                i.remove(number_input)
                shelf_input = input('Введите номер целевой полки')
                if shelf_input in directories.keys():
                    for k in directories.items():
                        k = list(k)
                        if shelf_input in k:
                            k[1].append(number_input)
                    return print('Документ перемещен')
                else:
                    print('Вы ввели номер несуществующей полки')
                    return move_doc()
    else: 
        print('Вы ввели номер несуществуемого документа')
        return move_doc()
def new_doc():
    type_input = input('Введите тип добавляемого документа')
    number_input = input('Введите номер добавляемого документа')
    name_input = input('Введите имя владельца добавляемого документа')
    shelf_input = input('Введите номер полки, в которую будет добавлен документ')
    if shelf_input in directories.keys():
        i = {'type': type_input, 'number': number_input, 'name': name_input}
        documents.append(i)
        for k in directories.items():
            k = list(k)
            if shelf_input in k:
                k[1].append(number_input)
        return print('Документ добавлен')
    else:
        print('Вы ввели номер несуществующей полки')
        return new_doc()
def show_list():
    for i in documents:
        print(*i.values())
print('Для просмотра всех доступных команд введите "help". Для завершения программы введите "q"')
while True:
    user_input = input('Введите команду ')
    if user_input == 'q':
        break
    else:
        main()
```
